# Read Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "/Users/gokcesoylu/bools.jpeg"
image = cv2.imread(img_path)
```

Display image

```python
plt.imshow(image)
plt.show()
```

## Convert red-green-blue to blue-greem-red

```python
bgr_image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
```

## displaying the image

```python
plt.imshow(bgr_image)
plt.show()
```

# Convert Gray

```python
gray_iamge = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_iamge,cmap='gray')
plt.show()
```

# Apply Threshold


```python
ret, tresholded_img = cv2.threshold(gray_iamge,127,255,cv2.THRESH_BINARY)
```

displaying

```python
plt.imshow(tresholded_img, cmap='gray')
plt.show()
```

# Find Contours

```python
contours, hierarchy = cv2.findContours(tresholded_img,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contour_image = cv2.cvtColor(tresholded_img, cv2.COLOR_GRAY2BGR)
```

```python
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
```

displaying

```python
plt.imshow(contour_image)
plt.show()
```

# Draw Bounding Box of Maximum Area Object 

en büyük alanı bulalım

```python
max_contour = max(contours, key=cv2.contourArea)
x, y, w, h = cv2.boundingRect(max_contour)
```

sınır kutusunu çizelim

```python
cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

displaying

```python
plt.imshow(image)
plt.title('Maximum Area')
plt.show()
```

# Crop the Maximum Area Object 

kırma işlemi

```python
cropped_object = image[y:y+h, x:x+w]
```

renklendirip görüntüleme işlemi

```python
plt.imshow(cv2.cvtColor(cropped_object, cv2.COLOR_BGR2RGB))
plt.title('Cropped Maximum Area Object')
plt.show()
```

# Write to File of the Maximum Area Object

```python
cv2.imwrite('cropped_object.jpg', cropped_object)
```
