# Read Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "/Users/gokcesoylu/uzum.jpeg"
image = cv2.imread(img_path)
```

Display image

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_RGB2BGR))
plt.show()
```

# Convert Gray

```python
gray_iamge = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_iamge,cmap='gray')
plt.show()
```
