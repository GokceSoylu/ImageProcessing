# Read Image

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img_path = "/Users/gokcesoylu/bools.jpeg"
image = cv2.imread(img_path)
```

Display image

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_RGB2BGR))
plt.show()
```

# Convert Gray

```python
gray_iamge = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_iamge,cmap='gray')
plt.show()
```

# Apply Threshold


```python
ret, tresholded_img = cv2.threshold(gray_iamge,127,255,cv2.THRESH_BINARY)
```

displaying

```python
plt.imshow(tresholded_img, cmap='gray')
plt.show()
```

# Find Contours

```python
contours, hierarchy = cv2.findContours(tresholded_img,cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contour_image = cv2.cvtColor(tresholded_img, cv2.COLOR_GRAY2BGR)
```

```python
cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
```

displaying

```python
plt.imshow(contour_image)
plt.show()
```

```python
# En büyük alanı bul
max_contour = max(contours, key=cv2.contourArea)
x, y, w, h = cv2.boundingRect(max_contour)

# Sınır kutusunu çiz
cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)

plt.imshow(image)
plt.title('Maximum Area')
plt.axis('off')
plt.show()
```
