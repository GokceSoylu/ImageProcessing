# Treshold

### cv2.treshold(src,tresh, maxval,type)

- src : verdiğmiz image. bu genellikle grayscale bi rgörüntüdür. yani öncelikle gray yaparız resmi

- tresh : belirlediğimiz eşik pixel. 

- maxval : belirlenen eşik değerin üzerindeki pixsellerin alacağı yeni değer.

- type : eşikelme türü.
- - cv2.THRESH_BINARY: Eşik değerinin üstündeki piksellere maxval atanır, altındakilere 0 atanır.
- - cv2.THRESH_BINARY_INV: Eşik değerinin altında olan piksellere maxval atanır, üstündekilere 0 atanır.
- - cv2.THRESH_TRUNC: Eşik değerinin üstündeki pikseller thresh değerine ayarlanır, altındakiler değişmez.
- - cv2.THRESH_TOZERO: Eşik değerinin altında olan piksellere 0 atanır, üstündekiler değişmez.
- - cv2.THRESH_TOZERO_INV: Eşik değerinin üstünde olan piksellere 0 atanır, altındakiler değişmez  

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img_path = "/Users/gokcesoylu/bardaklar.jpeg"
image = cv2.imread(img_path)
```

```python
image_bgr = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
```

```python
plt.imshow(image_bgr)
plt.show()
```

```python
image_bgr[295,300]
```

```python
ret, tresholded_img = cv2.threshold(image_bgr,132,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(tresholded_img)
plt.show()
```

```python
gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
plt.show()
```

```python
gray_img[300,280]
```

```python
ret, tresholded_img2 = cv2.threshold(gray_img,240,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(tresholded_img2,cmap='gray')
plt.show()
```
