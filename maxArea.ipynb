# Finding Maximum Area

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path = "/Users/gokcesoylu/bools.jpeg"
image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
arr, binary_img = cv2.threshold(img_gray,75,255,cv2.THRESH_BINARY)#bardak resmi iin 240 kullandk
contours, contours_array = cv2.findContours(binary_img,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(image, contours, -1,(255,0,0),2)
max_contour = max(contours, key=cv2.contourArea) #maksimum alana sahip kontur bulundu
x,y,w,h = cv2.boundingRect(max_contour) #max alan sahip kontrn bulnduu alan bi rrectangle olanarak belirlendi
cv2.rectangle(image, (x,y), (x+w, y+h),(255,255,255),2)
plt.imshow(image)
plt.show()
```
