# Convert Image to Gray

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## read image

```python
img_path = "/Users/gokcesoylu/lionpool.jpeg"
the_image = cv2.imread(img_path)
```

## convert grayscale

```python
gray_image = cv2.cvtColor(the_image, cv2.COLOR_BGR2GRAY)
```

## display image

```python
plt.imshow(gray_image,cmap='gray')
plt.show()
```
